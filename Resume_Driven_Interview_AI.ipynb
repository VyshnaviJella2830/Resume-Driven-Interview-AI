```python
import spacy
import random

# Load NLP model
nlp = spacy.load("en_core_web_sm")

print("🔥 AI Interview Question Generator Ready!")

# Sample Skill Keywords
skill_questions = {
    "python": [
        "Explain Python OOP concepts.",
        "What are decorators in Python?",
        "Difference between list and tuple?"
    ],
    "machine learning": [
        "What is overfitting?",
        "Explain bias vs variance.",
        "Difference between supervised and unsupervised learning?"
    ],
    "data analysis": [
        "How do you handle missing data?",
        "Explain data normalization.",
        "What is exploratory data analysis?"
    ],
    "html": [
        "Difference between HTML and HTML5?",
        "What are semantic tags?",
        "Explain forms in HTML."
    ],
    "css": [
        "What is Flexbox?",
        "Difference between inline and block elements?",
        "Explain CSS positioning."
    ],
    "javascript": [
        "What is closure in JavaScript?",
        "Difference between var, let, const?",
        "Explain event loop."
    ],
    "sql": [
        "Difference between INNER JOIN and LEFT JOIN?",
        "What is normalization?",
        "Explain indexing in SQL."
    ]
}

# Behavioral Questions
behavioral_questions = [
    "Tell me about yourself.",
    "Describe a challenging project you worked on.",
    "How do you handle tight deadlines?",
    "Explain a situation where you solved a difficult problem."
]

# User Resume Input
resume_text = input("\nPaste your Resume text here:\n").lower()

# Extract skills
detected_skills = []
for skill in skill_questions.keys():
    if skill in resume_text:
        detected_skills.append(skill)

print("\n📌 Detected Skills:", detected_skills)

print("\n🎯 Generated Interview Questions:\n")

# Generate Technical Questions
for skill in detected_skills:
    questions = random.sample(skill_questions[skill], min(2, len(skill_questions[skill])))
    print(f"🔹 {skill.upper()} Questions:")
    for q in questions:
        print("-", q)
    print()

# Add Behavioral Questions
print("🧠 Behavioral Questions:")
for q in random.sample(behavioral_questions, 2):
    print("-", q)
```
